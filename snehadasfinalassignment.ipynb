```python
#question2



import re


with open("aligned.fasta.txt","r") as myfile:
    
    align=myfile.read()

align=re.sub('>','delimiter>',align)

align=re.split('delimiter',align)

align=align[1:]

get_organism=re.compile('\[(\w+) (\w+)\]')

def_line=re.compile('(>.*)')



edited_list=[]

for fasta in align:
    
    organism=get_organism.search(fasta)
    
    abbre='>' +organism.group(1)[0]+'.'+organism.group(2)
    
    out=def_line.sub(abbre,fasta)
    
    edited_list.append(out)
    


with open('edited_fasta','w') as outfile:
    
    outfile.write(''.join(edited_list))
    
```
